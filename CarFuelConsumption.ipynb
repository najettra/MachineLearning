```python
df.drop(['refill gas', 'refill liters', 'specials'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
def repla(x):
    x=x.str.replace(',','.')
    return x

df[['distance','consume','temp_inside']]=df[['distance','consume','temp_inside']].apply(repla).astype(float)
```

```python
df.head()
```

```python
df.describe()
```

```python
df['temp_inside'].fillna(df['temp_inside'].mean(),inplace=True)
df.info()
```

```python
sb.pairplot(df,hue='gas_type')
```

```python
df.replace({'gas_type': {'E10':0, 'SP98':1}}, inplace=True)
```

```python
df.sample(10)
```

```python
cols = ['distance','consume','speed','temp_inside','temp_outside','gas_type','AC','rain','sun']
Y = df[['consume']]
X = df[['distance','speed','temp_inside','temp_outside','gas_type','AC','rain','sun']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test , y_train , y_test = train_test_split(X,Y,test_size=0.3,random_state=26)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
sb.distplot(y_test-predictions)
```

```python
from sklearn import metrics
```

```python
metrics.mean_squared_error(y_pred=predictions,y_true=y_test)
```

```python
metrics.explained_variance_score(y_pred=predictions,y_true=y_test)
```
